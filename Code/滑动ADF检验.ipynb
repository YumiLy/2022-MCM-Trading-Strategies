```python
# -*- encoding: utf-8 -*-
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
```

```python
# data preparing
BCHAIN = pd.read_csv('BCHAIN-MKPRU.csv', header=0)
GOLD = pd.read_csv('LBMA-GOLD.csv', header=0)
# to datetime
BCHAIN.Date = pd.to_datetime(BCHAIN.Date)
GOLD.Date = pd.to_datetime(GOLD.Date)
# merge
df = pd.merge(BCHAIN, GOLD, how='inner',on='Date')
df.columns = ['Date', 'BCHAIN', 'GOLD']
df = df.set_index('Date').rename_axis('Asset', axis=1)
# before fill
df.GOLD = df.GOLD.fillna(method="ffill")
df = df.dropna()
df.head()
```

```python
from statsmodels.tsa.stattools import adfuller
    
BCHAIN_adf = np.zeros(len(df)-5)
GOLD_adf = np.zeros(len(df)-5)
for i in range(0, len(df)-5):
    BCHAIN_adf[i] = adfuller(df.BCHAIN[i:i+5])[0]
    GOLD_adf[i] = adfuller(df.GOLD[i:i+5])[0]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index[:-5], y=BCHAIN_adf,
                    mode='markers',
                    name='BCHAIN'))
fig.add_trace(go.Scatter(x=df.index[:-5], y=GOLD_adf,
                    mode='markers',
                    name='GOLD',
                    line_color='green'))
fig.add_trace(go.Scatter(x=df.index[:-5], y=np.ones(len(df)-5),
                    mode='lines',
                    line=dict(color='red', width=3, dash='dash'),
                    name = '1'))
fig.add_trace(go.Scatter(x=df.index[:-5], y=-1*np.ones(len(df)-5),
                    mode='lines',
                    line=dict(color='red', width=3, dash='dash'),
                    name = '-1'))
```
