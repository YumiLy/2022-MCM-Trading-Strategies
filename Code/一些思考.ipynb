1、黄金和比特币数据分别的预测模型，能够拟合预测分别的走势、未来几天的交易情况

2、预测可能会有预测未来1天、3天、5天、7天，甚至更长，那么这个预测天数该如何选择，能够贴近实际，又能使得收益最大化。

3、为了使得整体收益最大，那么会考虑现金、黄金、比特币分别持有比例，以及在交易过程中的现金补仓，比特币卖出买入黄金，黄金卖出买入比特币，产品卖出换为现金等，所以是这三者相互交换的过程，需要考虑交换时机、交换数量、持有数量等之间的关系，形成一个动态的交易策略。

4、所形成策略需要有客观化的指标定义或阈值评判，当预测达到某些给定指标值或超过阈值，则实行交易策略。因此，所构建交易策略一定需要有客观的可执行的策略方案，通过一些给定指标来实现其客观化，可以照本宣科。

5、交易过程中会有手续费，那么不同的交易产生的利润需要结合手续费，短频快的方式会导致手续费上升，长期持有可能导致利润降低，需要有优化目标来衡量。

6、交易过程中不一定所有交易都能成功，有可能计划卖出比特币买黄金，但是卖出比特币后，黄金没有出货量，没办法或者没来得及以预计的价格买入。所以在交易过程中，可能出现比预定的卖出买入价格要不同，或高或低，更有可能交易失败。那么这些因素，不应该直接假设所有交易都可以按照模型设定都可以完成，而应该作为敏感性分析来进行讨论，交易策略也就变成一个区间策略，这样的话更能体现交易策略的容纳范围，并考虑交易过程的随机影响。一些交易失败概率，买入卖出价格的定价点和区间范围计算，可能都是需要的。

7、计算直接先要进行数据探索或预处理，看看是否有什么异常点、缺失等问题，先避免采用错误的数据或不完全的数据进行分析，一些必要的填补可能是需要的。异常值可能是有些特殊事件，导致价格波动，这类不一定适合去修正删除。数据不是独立的，数据是时间序列数据，那么一些时间序列建模方法可能可以采用，ARCH GARCH等模型可能可以考虑。

8、题目也要求考虑交易手续费对策略的影响，那么不同的手续费额度，会导致整个策略收益的不同和策略方案的不同，所以需要给出不同手续费下，策略的变化和收益的变化，从而方便研究手续费的策略的影响。

9、题目要求给出证据说明所提出策略是最优的，除了对策略的很好的模型阐述，分析原理，说明其合理性优越性以外，一般最好能有直观的证据，例如假设我就投资一种产品获得收益，我随机交叉投资的收益等等，通过一些可能的投资手段相互比较，从数值上说明本文所提出策略是最优的。当然，原理阐述和策略机制分析需要比较好的阐述所提出策略最优，虽然无法比较所有策略，但几种策略的相互比较是能够给出一定的直观证据的。
